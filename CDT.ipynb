```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
N = 10000
T = np.linspace(0,20,N)
dt = 10/N
S = np.zeros((N,4))
S[0] = [1e11,800e-9,450e-9, 800/450]
K0 = 30e-13/250e-9
K1 = 5e-13/250e-9
Ga = 0e-12
c = 0
d1 = 0 
d2 = 0
d0 = 0
for i in range(0 , N-1):
    d1 = S[i][0]/4*(-K0*S[i][1]+2*K1*S[i][2])*S[i][1] - 2*Ga*S[i][0]/3*(S[i][1]-S[i][2])+c
    d2 = S[i][0]/4*(K0*S[i][1])*S[i][2] + Ga*S[i][0]/3*(S[i][1]-S[i][2])+c
    d0 = - (K0*S[i][1]+2*K1*S[i][2]) * np.power(S[i][0],2)  - S[i][0]/2/S[i][1]*d1  - S[i][0]/S[i][2]*d2  
    S[i+1][0] =  S[i][0] + d0*dt 
    S[i+1][1] =  S[i][1] + d1*dt
    S[i+1][2] =  S[i][2] + d2*dt
    S[i+1][3] = S[i+1][1]/S[i+1][2]
```

```python
plt.figure()
#plt.plot(S[:,0],T)
plt.plot(T,S[:,1])
plt.plot(T,S[:,2])

plt.figure()
fig, ax1 = plt.subplots()

color = 'tab:brown'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('n', color=color)
ax1.plot(T, S[:,0], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Aspect Ratio', color=color)  # we already handled the x-label with ax1
ax2.plot(T, S[:,3], color='tab:red')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
